```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#A.Loading the dataset
data=pd.read_csv("student-mat.csv", sep=";")
```

```python
#B.Explore the data
print("Shape of dataset:",data.shape)
print("\nColumn names:", data.columns)
print("\nMissing values:\n", data.isnull().sum())
```

```python
#C. Clean the dataset
data=data.drop_duplicates()
data=data.fillna(data.median(numeric_only=True))
```

```python
#D
#Average final grade(G3)
print("Average G3:",data['G3'].mean())
```

```python
#D
#No.of students scoring above 15 in G3
print("Students with G3>15:", (data['G3']>15).sum())
```

```python
#D
#Correlation between study time and G3
print("Correlation between Study Time and G3:", data['studytime'].corr(data['G3']))
```

```python
#D
#Which gender has higher average G3?
average_by_gender=data.groupby('sex')['G3'].mean()
print("Average by gender:\n", average_by_gender)
print("Gender with higher average G3:", average_by_gender.idxmax())
```

```python
#E
#Histogram
plt.hist(data['G3'],bins=10,edgecolor='black')
plt.xlabel("Final Grade(G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades(G3)")
plt.show()
```

```python
#E
#Scatter plot
plt.scatter(data['studytime'],data['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade(G3)")
plt.title("Study Time vs Final Grade(G3)")
plt.show()
```

```python
#E
#Bar Chart
sns.barplot(x='sex', y='G3', data=data, errorbar=None)
plt.title("Average Final Grade (G3) by Gender")
plt.show()
```
